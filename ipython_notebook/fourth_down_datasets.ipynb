```python
import pandas as pd
import numpy as np
import math
import csv
```

```python
pd.set_option('display.line_width', 300)
df = pd.read_csv('FourthDownPlays.csv')
df.head(2)
```

```python
cond = df['description'].str.contains('field goal' or 'Field Goal' or 'field Goal' or 'Field goal' or 'FIELD GOAL')
field_goal_data = df[cond == True]
field_goal_data.head(2)
```

```python
pbp = pd.read_csv('all_season_pbp_data.csv')
pbp.head(1)
```

```python
play_after_field_goal = pbp.ix[field_goal_data['Unnamed: 0'] + 1, :]
play_after_fg_important_vars = play_after_field_goal[['offscore', 'defscore', 'off']]
play_after_fg_important_vars.head(2)
```

```python
play_after_fg_important_vars.columns = ['offscore2', 'defscore2', 'off2']
play_after_fg_important_vars.index = field_goal_data.index
field_goal_data_2 = field_goal_data.join(play_after_fg_important_vars)
field_goal_data_2.head(2)
```

```python
field_goal_data_2['converted'] = ((field_goal_data_2['off'] == field_goal_data_2['off']) & 
                                (field_goal_data_2['offscore2'] == field_goal_data_2['offscore'] + 3))
field_goal_data_2['distance'] = field_goal_data_2['ydline'] + [18]*len(field_goal_data_2['ydline'])
field_goal_data_final = field_goal_data_2[['ydline', 'distance', 'description', 'converted']]
field_goal_data_final.head()
```

```python
cond2 = df['description'].str.contains('punt' or 'PUNT' or 'Punt')
punt_data = df[cond2 == True]
punt_data.head(2)
```

```python
punt_data['ydlineotherteam'] = 100 - (punt_data['ydline']-38)
punt_data_final = punt_data[['description', 'ydline', 'ydlineotherteam']]
punt_data_final.head()
```

```python
cond = df['description'].str.contains('field goal' or 'Field goal' or 'field Goal' or 'Field Goal' or 'FIELD GOAL')
go = df[cond == False]
cond2 = go['description'].str.contains('punt' or 'PUNT' or 'Punt')
go_for_it = go[cond2 == False]
go_for_it.head(2)
```

```python
play_after_going_for_it = pbp.ix[go_for_it['Unnamed: 0'] + 1, :]
play_after_gfi_important_vars = play_after_going_for_it[['ydline', 'offrank', 'off']]
play_after_gfi_important_vars.head(2)
```

```python
play_after_gfi_important_vars.columns = ['ydline2', 'offrank2', 'off2']
play_after_gfi_important_vars.index = go_for_it.index
go_for_it_2 = go_for_it.join(play_after_gfi_important_vars)
go_for_it_2.head(1)
```

```python
go_for_it_2['converted'] = go_for_it_2['off'] == go_for_it_2['off2']
go_for_it_final = go_for_it_2[['offrank', 'defrank', 'togo', 'converted']]
go_for_it_final.head()
```
