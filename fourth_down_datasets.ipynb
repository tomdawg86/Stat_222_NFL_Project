```python
import pandas as pd
import numpy as np
import math
import csv
import scipy as sp
import statsmodels.api as sm
```

```python
pd.set_option('display.line_width', 300)
df = pd.read_csv('FourthDownPlays.csv')
df.head(2)
```

```python
cond = df['description'].str.contains('field goal' or 'Field Goal' or 'field Goal' or 'Field goal' or 'FIELD GOAL')
field_goal_data = df[cond == True]
field_goal_data.head(2)
```

```python
field_goal_data['converted'] = field_goal_data['description'].str.contains('GOOD')
field_goal_data['distance'] = field_goal_data['ydline'] + 18
field_goal = field_goal_data[['ydline', 'distance', 'description', 'converted']]  
field_goal.head()
```

```python
field_goal['intercept'] = [1]*len(field_goal)
field_goal_logit = sm.Logit(field_goal[['converted']], field_goal[['intercept','ydline']])
field_goal_result = field_goal_logit.fit()
field_goal_result.summary()
```

```python
field_goal_result.predict([1, 30], True)
```

```python
cond2 = df['description'].str.contains('punt' or 'PUNT' or 'Punt')
punt_data = df[cond2 == True]
punt_data.head(2)
```

```python
go = df[cond == False]
cond2 = go['description'].str.contains('punt' or 'PUNT' or 'Punt')
go_for_it = go[cond2 == False]
go_for_it.head(2)
```

```python
pbp = pd.read_csv('all_season_pbp_data.csv')
pbp.head(1)
```

```python
play_after_going_for_it = pbp.ix[go_for_it['Unnamed: 0'] + 1, :]
play_after_gfi_important_vars = play_after_going_for_it[['ydline', 'offrankbucket']]
play_after_gfi_important_vars.head(2)
```

```python
play_after_gfi_important_vars.columns = ['ydline2', 'offrankbucket2']
play_after_gfi_important_vars.index = go_for_it.index
go_for_it_2 = go_for_it.join(play_after_gfi_important_vars)
go_for_it_2['converted'] = go_for_it_2['offrankbucket'] == go_for_it_2['offrankbucket2']
go_for_it_2.head(2)
```

```python
go_for_it_final = go_for_it_2[['offrankbucket', 'defrankbucket', 'togo0t1', 'togo2t3', 'togo4t5', 'togo6t7', 'togo8t10', 'togo11t13', 'togo14t16', 'togo17t20', 'offrank1t5', 'offrank6t10', 'offrank11t15', 'offrank16t20', 'offrank21t25', 'defrank1t5', 'defrank6t10', 'defrank11t15', 'defrank16t20', 'defrank21t25', 'converted']]
go_for_it_final.head()
```

```python
go_for_it_final['intercept'] = [1]*len(go_for_it_final)
go_for_it_logit = sm.Logit(go_for_it_final[['converted']], go_for_it_final[['intercept','togo0t1', 'togo2t3', 'togo4t5', 'togo6t7', 'togo8t10', 'togo11t13', 'togo14t16', 'togo17t20', 'offrank1t5', 'offrank6t10', 'offrank11t15', 'offrank16t20', 'offrank21t25', 'defrank1t5', 'defrank6t10', 'defrank11t15', 'defrank16t20', 'defrank21t25']])
go_for_it_result = go_for_it_logit.fit()
go_for_it_result.summary()
```
